```python
new_series = students.marks + 5
new_series
```

```python
new_series.map(lambda mark: "Pass" if mark >= 70 else "Needs Improvement")
```

```python
students["adjusted_marks"] = students.marks.map(lambda m: m + 5)
students
```

```python
students["status"] = students.marks.map(lambda mark: "Pass" if mark >= 70 else "Needs Improvement")
students
```

```python
def effort_label(row):
    if row.hours_studied >= 4:
        return "High Effort"
    else:
        return "Low Effort"

students["Effort"] = students.apply(effort_label, axis = "columns")
students
```

## 4. Grouping and Sorting

```python
students
```

```python
students.groupby(["status"]).size()
```

```python
students.groupby(["status"]).marks.mean()
```

```python
students.groupby(["status"]).hours_studied.mean()
```

```python
students.groupby(["status"]).marks.agg(["max", "min"])
```

```python
students.groupby(["Effort"]).size()
```

```python
students.groupby(["Effort"]).marks.mean()
```

```python
students.sort_values(by = ["marks"])
```

```python
students.sort_values(by = ["marks"], ascending= False)
```

```python
students.sort_values(by = ["hours_studied"], ascending= False)
```

```python
students.sort_values(by = ["status", "marks"])
```

```python
students.sort_values(by = ["Effort", "marks"], ascending= [True, False])
```

```python
students_missing = pd.DataFrame({
    "name": ["A", "B", "C", None, "E"],
    "marks": [82, None, 76, 65, None],
    "hours_studied": [4, 5, None, 2, 1]
})

students_missing
```

```python
students_missing.isnull()
```

```python
students_missing.isnull().sum(axis = "index")
```

```python
students_missing[students_missing.marks.isnull()]
```

```python
students_missing.marks = students_missing.marks.fillna(0)
```

```python
students_missing.name = students_missing.name.fillna("Unknown")
students_missing
```

```python
students_missing.hours_studied = students_missing.hours_studied.fillna(students_missing.hours_studied.mean())
students_missing
```

## 6. Renaming and Combining

```python

students_renamed = students.rename(columns={
    "name": "student_name",
    "marks": "original_marks",
    "hours_studied": "study_hours"
})

students_renamed
```

```python

new_students = pd.DataFrame({
    "name": ["F", "G"],
    "marks": [88, 72],
    "hours_studied": [4, 3],
    "adjusted_marks": [93, 77],
    "status": ["Pass", "Pass"],
    "Effort": ["High Effort", "Low Effort"]
})

all_students = pd.concat([students, new_students], ignore_index=True)
all_students
```

```python
student_goals = pd.DataFrame({
    "name": ["A", "B", "C", "D", "E"],
    "goal": ["AI", "Data Analysis", "Machine Learning", "Web Development", "Cybersecurity"]
})

students_with_goals = students.merge(student_goals, on="name", how="left")
students_with_goals
```

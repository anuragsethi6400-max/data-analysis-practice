# Titanic Data Analysis

A beginner exploratory data analysis project using the Kaggle Titanic training dataset.

```python
import pandas as pd

df = pd.read_csv("train.csv")
df.head()
```

## Analysis Questions

1. How many rows and columns are in the dataset?

```python
df.shape
```

rows = 891
columns = 12

2. What are all column names?

```python
df.columns
```

3. Which columns have missing values, and how many?

```python
df.isnull().sum()
```

Age(177), Cabin(687) and Embarked(2) Columns have missing values.

4. What is the average passenger age?

```python
df.Age.mean().round(2)
```

The average Age is 29.7 years

5. How many passengers survived?

```python
df.Survived.sum()
```

Total 342 passengers survived.

6. What percentage of passengers survived?

(df.Survived.sum()/len(df) * 100).round(2)

38.38 % passengers survived

7. What is the survival rate for female versus male passengers?

```python
df.groupby("Sex").Survived.mean() * 100
```

8. What is the survival rate for each passenger class (Pclass)?

```python
df.groupby("Pclass").Survived.mean() * 100
```

9. What is the average fare for each passenger class?

```python
df.groupby("Pclass").Fare.mean()
```

10. Who was the oldest passenger and who was the youngest passenger?

```python
df.Age.agg(['max', 'min'])
```

11. Does survival change by Embarked location?

```python
df.groupby("Embarked").Survived.mean() * 100
```

Embarked = 'C' has the highest survival rate.

12. Compare survival for women in first, second, and third class.

```python
df.groupby(["Sex", "Pclass"]).Survived.mean()["female"]
```

## Key Findings

- Finding 1: More Females survived more than the males
- Finding 2: Fare of the 1-Class is much high as compared to 2 and 3 classes
- Finding 3: Ages from 80 to 0.42 were onboarded

## 2. Functions

```python
def greeting(name):
    return f"Hello {name}"

def total_hours(hours_per_day, days):
    return hours_per_day * days

def c_to_f(temp_c):
    return (temp_c * 9/5) + 32

def check_if_pass(marks):
    if (marks >= 70):
        return "Pass"
    else:
        return "Needs Improvement"

def larger_num (a, b):
    if (a > b):
        return a
    elif (b > a):
        return b
    else:
        return "They are equal" 

print(greeting("Anurag"))
print(total_hours(3, 21))
print(c_to_f(0))
print(check_if_pass(82))
print(larger_num(10, 5))
```

## 10. Mini Student Marks Analysis

```python
students = [
    {"name": "Anne", "marks": 82},
    {"name": "Bob", "marks": 91},
    {"name": "Caven", "marks": 76},
    {"name": "Dobby", "marks": 65},
    {"name": "Ele", "marks": 58}
]

for dict in students:
    print(dict["name"])
```

```python
total_marks = 0
for dict in students:
    total_marks += dict["marks"]

average_marks = total_marks/len(students)
average_marks
```

```python
highest_scorer = students[0]["name"]
high_score = students[0]["marks"]

for dict in students:
    if (dict["marks"] > high_score):
        high_score = dict["marks"]
        highest_scorer = dict["name"]
highest_scorer
```

```python
lowest_scorer = students[0]["name"]
low_score = students[0]["marks"]

for dict in students:
    if (dict["marks"] < low_score):
        low_score = dict["marks"]
        lowest_scorer = dict["name"]
lowest_scorer
```

```python
above_80 = [s["name"] for s in students if s["marks"] > 80]
above_80
```

```python
for dict in students:
    if (dict["marks"] > 70):
        dict["pass"] = True
    else:
        dict["pass"] = False

students
```

```python
passed_students = sum(s["pass"] == True for s in students)
passed_students
```

```python
need_improvement = sum(s["pass"] == False for s in students)
need_improvement
```

```python
def get_grade(marks):
    if marks >= 85:
        return "A"
    elif marks >= 70:
        return "B"
    elif marks >= 60:
        return "C"
    else:
        return "D"

for student in students:
    student["grade"] = get_grade(student["marks"])

students
```
